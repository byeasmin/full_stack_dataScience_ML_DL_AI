#  Python — Pass by Object Reference

> Python does **not** use pure pass by value or pure pass by reference.
> It uses something called **"Pass by Object Reference"** *(also called "Pass by Assignment")*.

## What Does That Mean?

When we pass a variable to a function, Python sends the **reference of the object**.

- If the object is **mutable** (`list`, `dict`, etc.) → it **can** be changed inside the function.
- If the object is **immutable** (`int`, `str`, `tuple`) → it **cannot** be changed permanently.

## Immutable Example (`int`)

```python
def change(x):
    x = 100
a = 10
change(a)

print(a)
```
**Output:**
```
10
```
`int` is **immutable**, so the original value does **not** change.

## Mutable Example (`list`)

```python
def change(lst):
    lst[0] = 100

numbers = [1, 2, 3]
change(numbers)

print(numbers)
```

**Output:**
```
[100, 2, 3]
```
`list` is **mutable**, so the original list **does** change.

## Summary:

| Concept | Python's Behavior |
|---|---|
| Pass by Value |  Not exactly |
| Pass by Reference |  Not exactly |
| Pass by Object Reference |  Yes! |

> Python is officially **neither** pure pass by value **nor** pure pass by reference.
> It uses **pass by object reference**.

# Immutable Object

````python
def change(x):
    x = 100
    print(x)

a = 10
change(a)
print("a = ", a)
````

# Mutable Object

````python
def change(lst):
    lst[0] = 100
    print(lst)

numbers = [1, 2, 3]
change(numbers)

print("numbers = ",numbers)
````

# By ID

````python
def update(lst):
  print(id(lst))
  lst[0] = 100
  print(lst)
  print(id(lst))

a = [1, 2, 3]
print(id(a))
update(a)
````
