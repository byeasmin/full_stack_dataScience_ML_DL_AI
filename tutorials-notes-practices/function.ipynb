# return vs print (with example)



*   return gives a value back from a function so it can be stored, reused, or used in other calculations. It is used in most programs.




```python
a = subs(5, 4)
print(a)
```



*   print only shows output on the screen and does not give any value back for further use.



```python
a = add(5, 4)
print(a)
```

# Multiple Task

```python
def add_sub(m, n):
  c = m + n
  d = m - n
  return c, d
r1, r2 = add_sub(6, 8)
```

```python
print(r1, r2)
```

**But it always a good practice to have a seperate function for seperate task.**
