# Lists

A **list** is an ordered, mutable collection that can hold items of different types. Lists are the most versatile and commonly used data structure in Python.

## Creating Lists

````python
mixed = [1, "Hello", 3.14, True]
print(mixed)
````

`list()` constructor can be used with any **iterable** - strings, tuples, sets, ranges, dictionaries (gives keys), generators, and more. If you can loop through it with `for`, you can convert it to a list!

````python
from_string = list("Python")
print(f"From string: {from_string}")
````

````python
range_list = list(range(1, 6))
print(f"From range: {range_list}")
````

## Indexing and Slicing

Lists are **indexed**, meaning each element has a position number starting from 0.

````python
fruits = ["apple", "cherry", "date", "elderberry", "banana"]

print(fruits[0])
````

````python
print(fruits[-1])
````

````python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
# Slicing [start:stop:step]

# print(fruits)
````

````python
print(f"First 3 fruits: {fruits[0:3]}")
print(f"From index 2 to end: {fruits[2:]}")
print(f"Last 2 fruits: {fruits[-2:]}")
print(f"Every 2nd fruit: {fruits[::2]}")
print(f"Reverse list: {fruits[::-1]}")
````

## List Methods

Lists have many built-in methods for manipulation.

### 1. `append()` - Add element at the end

````python
fruits = ["apple", "banana"]
fruits.append("cherry")
print(f"After append: {fruits}")
````

### 2. `insert()` - Add element at specific position

````python
fruits = ["apple", "cherry"]
fruits.insert(1, "banana")
print(f"After insert: {fruits}")
````

### 3. `remove()` - Remove first occurrence of a value

````python
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")
print(f"After remove: {fruits}")
````

````python
# fruits.remove("mango")  # ❌
````

### 4. `pop()` - Remove and return element at index (default: last)

````python
fruits = ["apple", "banana", "cherry", "date"]

last = fruits.pop()
print(f"Popped: {last}")
print(f"Remaining: {fruits}")
````

````python
second = fruits.pop(1)
print(f"Popped at index 1: {second}")
print(f"Remaining: {fruits}")
````

### 5. `clear()` - Remove all elements

````python
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(f"After clear: {fruits}")
````

### 6. `index()` - Find position of first occurrence

````python
fruits = ["apple", "banana", "cherry", "banana"]

x = "cherry"
position = fruits.index(x)
print(f"Position of '{x}': {position}")
````

````python
# fruits.index("mango")  # ❌ 
````

### 7. `count()` - Count occurrences of a value

````python
numbers = [1, 2, 3, 2, 4, 2, 5]

k = 2

count = numbers.count(k)
print(f"Count of {k}: {count}")
````

### 8. `sort()` - Sort list in place

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]


numbers.sort()
print(f"Sorted ascending: {numbers}")
````

````python
numbers.sort(reverse=True)
print(f"Sorted descending: {numbers}")
````

````python
# Sort strings alphabetically
fruits = ["cherry", "apple", "banana"]
fruits.sort()
print(f"Sorted fruits: {fruits}")
````

### 9. `reverse()` - Reverse list in place

````python
numbers = [1, 2, 3, 4, 5]
numbers.reverse()
print(f"Reversed: {numbers}")
````

````python
# To get reversed copy without modifying: numbers[::-1]
print(f"before {numbers}")
print(f"reversed {numbers[::-1]}")
print(f"after {numbers}")
````

### 10. `copy()` - Create a shallow copy

````python
original = [1, 2, 3]

# Wrong way - creates reference
copy = original
copy.append(4)
print(f"Original: {original}")
print(f"Reference: {copy}")
````

````python
original = [1, 2, 3]
copy = original.copy()
copy.append(4)
print(f"Original: {original}")
print(f"Copy: {copy}")
````

````python
squares = [x**2 for x in range(10)]
print(f"Squares: {squares}")
````

````python
evens = [x for x in range(20) if x % 2 == 0]
print(f"Even numbers: {evens}")
````

# Tuples

A **tuple** is an ordered, **immutable** collection. Once created, you cannot modify it (no adding, removing, or changing elements).

## Creating Tuples

````python
coordinates = (23.8103, 90.4125)
print(f"Coordinates: {coordinates}")
````

````python
rgb_color = 255, 0, 0  # Red
print(f"RGB Color: {rgb_color}")
````

````python
empty = ()
print(f"Empty tuple: {empty}")

mixed = (1, "Hello", 3.14, True)
print(f"Mixed tuple: {mixed}")
````

## Accessing Tuple Elements

Same as lists - using indexing and slicing.

````python
fruits = ("apple", "banana", "cherry", "date")

# Indexing
print(f"First: {fruits[0]}")

# Slicing
print(f"First two: {fruits[0:2]}")
````

````python
coordinates = (23.8103, 90.4125)

# coordinates[0] = 24.0

# coordinates.append(100)

# Tuples are immutable - cannot be changed after creation!
````

````python
numbers = (1, 4, 2, 3, 2, 4, 2, 5, 4)

count_of_2 = numbers.count(2)
print(f"Count of 2: {count_of_2}")

position = numbers.index(4)
print(f"Position of 4: {position}")
````

## Tuple Unpacking

Extract values from tuple into separate variables.

````python
color = (255, 0, 0)
red, green, blue = color
print(f"Red: {red}, Green: {green}, Blue: {blue}")
````

````python
# Using * to collect remaining values
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(f"First: {first}, Middle: {middle}, Last: {last}")
````

````python
# Example: Function returning multiple values
def get_student_info():
    name = "Ayan"
    age = 7
    grade = "A"
    return name, age, grade

student_name, student_age, student_grade = get_student_info()
print(f"Name: {student_name}, Age: {student_age}, Grade: {student_grade}")
````

# Sets

A **set** is an **unordered** collection of **unique** elements. Sets automatically remove duplicates and are mutable.

## Creating Sets

````python
fruits = {"cherry", "banana", "cherry", "apple"}
print(f"Fruits set: {fruits}")
````

````python
# Empty set - must use set(), not {}
empty = set()  # Correct
not_empty_set = {}  # This creates an empty dictionary!
print(f"Empty set: {empty}, Type: {type(empty)}")
print(f"Empty dict: {not_empty_set}, Type: {type(not_empty_set)}")
````

````python
chars = set("hello")
print(f"Unique characters in 'hello': {chars}")
````

````python
numbers = {1, 2, 3, 2, 1} 

# print(numbers[0])
````

## Set Methods

### Adding and Removing Elements

````python
fruits = {"apple", "banana"}

fruits.add("cherry")
print(f"After add: {fruits}")
````

````python
fruits.add("apple")
print(f"After adding duplicate: {fruits}")
````

````python
fruits.remove("banana")
print(f"After remove: {fruits}")
````

````python
# fruits.remove("mango")
````

````python
fruits.discard("cherry")
print(f"After discard: {fruits}")
fruits.discard("mango") 
print(f"After discarding non-existent: {fruits}")
````

## Set Operations

Sets support mathematical set operations like union, intersection, and difference.

````python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

union = set_a.union(set_b)
print(f"Union: {union}")
````

````python
intersection = set_a.intersection(set_b) 
print(f"Intersection: {intersection}")
````

````python
difference = set_a.difference(set_b)
print(f"Difference (A - B): {difference}")
````

````python
# Symmetric difference - elements in either but not both
sym_diff = set_a.symmetric_difference(set_b)
print(f"Symmetric difference: {sym_diff}")
````

# Frozenset

A `frozenset` is an **immutable version** of a set. Once created, you cannot add or remove elements.

## Why Frozenset?

Since `frozenset` is **immutable**, it is **hashable** and can be used as:
- Dictionary keys
- Elements in other sets

## Creating Frozensets

````python
fs = frozenset([1, 2, 3, 2, 1])
print(f"Frozenset: {fs}")
````

````python
regular_set = {4, 5, 6}
fs2 = frozenset(regular_set)
print(f"Frozenset from set: {fs2}")
````

````python
fs3 = frozenset("hello")
print(f"Frozenset from string: {fs3}")
````

````python
fs = frozenset([1, 2, 3])

# fs.add(4)  

# fs.remove(1)
````

````python
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])
union = fs1 | fs2
print(f"Union of frozensets: {union}")
````

A **dictionary** is an unordered collection of **key-value pairs**. Each key is unique and maps to a value.

## Iterating Through Dictionaries

````python
student = {"name": "Ayan", "age": 7, "grade": "A", "city": "Dhaka"}

# Iterate over keys
print("Keys:")
for key in student:
    print(f"  {key}")
````

````python
# Iterate over key-value pairs
print("\nKey-Value pairs:")
for key, value in student.items():
    print(f"  {key}: {value}")
````

## Practical Example

````python
text = "hello world hello python world"
word_count = {}

for word in text.split():
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(f"Word frequency: {word_count}")
````

# Generators

A **generator** in Python is a special type of `iterable` that allows you to iterate over a sequence of values. Unlike lists, which store all values in memory, a generator computes each value **on the fly** and yields it one at a time, which makes it more **memory efficient**, especially for large datasets.

## Why Generators?

- **Memory Efficient** - Don't store all values at once
- **Lazy Evaluation** - Compute values only when needed
- **Infinite Sequences** - Can represent infinite data streams
- **Pipeline Processing** - Chain operations efficiently

## Generator Expressions

A generator expression is similar to a list comprehension, but it uses **parentheses `()`** instead of square brackets `[]`.

````python
# List comprehension - creates entire list in memory
squares_list = [x**2 for x in range(10)]
print(f"List comprehension: {squares_list}")
print(f"Type: {type(squares_list)}")
````

````python
# Generator expression - creates generator object
squares_gen = (x**2 for x in range(10))
print(f"Generator expression: {squares_gen}")
print(f"Type: {type(squares_gen)}")
````

````python
# Converting generator to list
squares_from_gen = list(squares_gen)
print(f"Generator converted to list: {squares_from_gen}")

# Once consumed, generator is empty
print(f"Generator after conversion: {list(squares_gen)}")
````

## Memory Efficiency Comparison

````python
import sys

large_list = [x for x in range(10000)]
print(f"List size: {sys.getsizeof(large_list):,} bytes")

large_gen = (x for x in range(10000))
print(f"Generator size: {sys.getsizeof(large_gen):,} bytes")

print(f"\nMemory saved: {sys.getsizeof(large_list) - sys.getsizeof(large_gen):,} bytes")
````

## `loc` vs `iloc` - What's the difference?

**`loc`** = Access by **LABEL/NAME**
- `df.loc[0]` → Row where index label is 0
- `df.loc[0, 'Name']` → Cell at index 0, column 'Name'

**`iloc`** = Access by **INTEGER POSITION** (like list indexing)
- `df.iloc[0]` → First row (position 0)
- `df.iloc[0, 0]` → First row, first column (position 0, 0)

**When index is 0,1,2,3... they give same result, but:**
```python
# With custom index
df = pd.DataFrame({'Name': ['Ayan', 'Rafi']}, index=[10, 20])

df.loc[10]   # ✅ Row with label 10 (first row)
df.iloc[0]   # ✅ Row at position 0 (first row)
df.loc[0]    # ❌ KeyError - no label 0
```

**Simple rule:** 
- Use **`loc`** when working with row/column names
- Use **`iloc`** when working with positions (like `list[0]`)
