## Problem 1: Student Class

Create a `Student` class with the following specifications:

**Attributes:**
- `name` (string)
- `age` (integer)
- `student_id` (string)
- `grades` (list, initialized as empty list)

**Methods:**
- `add_grade(grade)` - adds a grade to the grades list
- `get_average()` - returns the average of all grades (return 0 if no grades)
- `get_info()` - prints: "Name: [name], Age: [age], ID: [student_id]"

**Example:**
```python
student = Student("Alice", 20, "S001")
student.add_grade(85)
student.add_grade(90)
print(student.get_average())  # Output: 87.5
```

````python
class Student:
    def __init__(self, name, age, student_id):
        self.name = name
        self.age = age
        self.student_id = student_id
        self.grades = []
    def add_grade(self, grade):
        self.grades.append(grade)
    def get_average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)
    def get_info(self):
        print(f"Name: {self.name}, Age: {self.age}, ID: {self.student_id}")
````

## Problem 2: BankAccount Class

Create a `BankAccount` class with encapsulation:

**Attributes:**
- `account_number` (string)
- `__balance` (private float, initialized with initial_balance)

**Methods:**
- `deposit(amount)` - adds amount to balance, prints "Deposited: [amount]"
- `withdraw(amount)` - subtracts amount from balance if sufficient funds, prints "Withdrawn: [amount]"
  - If insufficient funds, print "Insufficient balance!" and don't withdraw
- `get_balance()` - returns the current balance

**Example:**
```python
account = BankAccount("ACC001", 1000)
account.deposit(500)  # Balance: 1500
account.withdraw(200)  # Balance: 1300
print(account.get_balance())  # Output: 1300
```

````python
class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.account_number = account_number
        self.__balance = initial_balance
    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance!")
    def get_balance(self):
        return self.__balance
````

````python
# Test Cell 
account = BankAccount("ACC001", 1000)
account.deposit(500)
account.withdraw(200)

account.withdraw(2000)
````

````python
account = BankAccount("ACC001", 1000)
account.deposit(500)  
account.withdraw(200)  
print(account.get_balance()) 
````

## Problem 3: DataPreprocessor Class

Create a `DataPreprocessor` class:

**Attributes:**
- `mean` (initialized to None)
- `std` (initialized to None)
- `__is_fitted` (private boolean, initialized to False)

**Methods:**
- `fit(data)` - calculates mean and std from data list, sets __is_fitted to True
- `transform(data)` - normalizes data using formula: (x - mean) / std
  - Returns normalized list
  - If not fitted, print "Error: Not fitted!" and return None
- `is_ready()` - returns the __is_fitted status

**Formulas:**
- mean = sum(data) / len(data)
- std = sqrt(sum((x - mean)^2 for x in data) / len(data))

**Example:**
```python
preprocessor = DataPreprocessor()
preprocessor.fit([10, 20, 30, 40, 50])
result = preprocessor.transform([20, 40])
print(result)  # Normalized values
```

````python
import math

class DataPreprocessor:
    def __init__(self):
        self.mean = None
        self.std = None
        self.__is_fitted = False
    def fit(self, data):
        self.mean = sum(data) / len(data)
        variance = sum((x - self.mean) ** 2 for x in data) / len(data)
        self.std = math.sqrt(variance)
        self.__is_fitted = True
    def transform(self, data):
        if not self.__is_fitted:
            print("Error: Not fitted!")
            return None
        if self.std == 0:
            return [0 for _ in data]
        return [(x - self.mean) / self.std for x in data]
    def is_ready(self):
        return self.__is_fitted
````

````python
# Test Cell
preprocessor = DataPreprocessor()
result = preprocessor.transform([20, 40])
preprocessor.fit([10, 20, 30, 40, 50])
result = preprocessor.transform([30])
````

````python
preprocessor = DataPreprocessor()
preprocessor.fit([10, 20, 30, 40, 50])
result = preprocessor.transform([20, 40])
print(result)  
````

## Problem 4: Person → Employee Inheritance

Create two classes with inheritance:

**Parent Class: Person**
- Attributes: `name`, `age`
- Method: `introduce()` - prints "Hi, I'm [name], [age] years old"

**Child Class: Employee (inherits from Person)**
- Additional Attributes: `employee_id`, `salary`
- Method: `work()` - prints "[name] is working..."
- Method: `get_details()` - prints "Name: [name], Age: [age], Employee ID: [employee_id], Salary: [salary]"

**Example:**
```python
emp = Employee("Bob", 30, "E001", 50000)
emp.introduce()  # Inherited method
emp.work()       # New method
```

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hi, I'm {self.name}, {self.age} years old")
class Employee(Person):
    def __init__(self, name, age, employee_id, salary):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.salary = salary
    def work(self):
        print(f"{self.name} is working...")
    def get_details(self):
        print(f"Name: {self.name}, Age: {self.age}, Employee ID: {self.employee_id}, Salary: {self.salary}")
````

````python
# Test Cell
person = Person("Alice", 25)
person.introduce()

emp = Employee("Bob", 30, "E001", 50000)
emp.introduce()
emp.work()
emp.get_details()
````

````python
emp = Employee("Bob", 30, "E001", 50000)
emp.introduce() 
emp.work()       
````

## Problem 5: StudentGradeSystem 

Create a `StudentGradeSystem` class:

**Attributes:**
- `system_name` (string)
- `__grades` (private dictionary, initialized as empty: {})

**Methods:**
- `add_student(student_name, grade)`
  - Validates: grade must be between 0 and 100
  - If invalid, print "Error: Grade must be between 0 and 100!" and don't add
  - Otherwise, add to __grades dictionary
  - Print "Added: [student_name] with grade [grade]"

- `get_grade(student_name)`
  - Returns grade for student
  - If student not found, return "Error: Student not found!"
  - Handle KeyError exception

- `get_class_average()`
  - Returns average of all grades
  - If no students, return "Error: No students in system!"
  - Handle division by zero

- `get_student_count()`
  - Returns number of students

**Example:**
```python
system = StudentGradeSystem("Math Class")
system.add_student("Alice", 85)
system.add_student("Bob", 92)
print(system.get_grade("Alice"))  # Output: 85
print(system.get_class_average())  # Output: 88.5
```

````python
class StudentGradeSystem:
    def __init__(self, system_name):
        self.system_name = system_name
        self.__grades = {}
    def add_student(self, student_name, grade):
        if grade < 0 or grade > 100:
            print("Error: Grade must be between 0 and 100!")
            return
        self.__grades[student_name] = grade
        print(f"Added: {student_name} with grade {grade}")
    def get_grade(self, student_name):
        try:
            return self.__grades[student_name]
        except KeyError:
            return "Error: Student not found!"
    def get_class_average(self):
        try:
            return sum(self.__grades.values()) / len(self.__grades)
        except ZeroDivisionError:
            return "Error: No students in system!"
    def get_student_count(self):
        return len(self.__grades)
system = StudentGradeSystem("Math Class")
system.add_student("Alice", 85)
system.add_student("Bob", 92)
print(system.get_grade("Alice"))  
print(system.get_class_average())
````

````python
# Test Cell

system = StudentGradeSystem("Math Class")

system.add_student("Alice", 85)
system.add_student("Bob", 92)
system.add_student("Charlie", 78)
print(f"Total students: {system.get_student_count()}")
print(f"Alice's grade: {system.get_grade('Alice')}")
print(f"Bob's grade: {system.get_grade('Bob')}")

system.add_student("David", 150)
print(f"Total students after invalid: {system.get_student_count()}")

result = system.get_grade("Unknown")
print(f"Result: {result}")

avg = system.get_class_average()
print(f"Class average: {avg}")

empty_system = StudentGradeSystem("Empty Class")
result = empty_system.get_class_average()
print(f"Empty system average: {result}")
````
