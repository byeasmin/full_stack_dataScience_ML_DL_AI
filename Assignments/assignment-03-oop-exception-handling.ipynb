## Problem 3: DataPreprocessor Class

Create a `DataPreprocessor` class:

**Attributes:**
- `mean` (initialized to None)
- `std` (initialized to None)
- `__is_fitted` (private boolean, initialized to False)

**Methods:**
- `fit(data)` - calculates mean and std from data list, sets __is_fitted to True
- `transform(data)` - normalizes data using formula: (x - mean) / std
  - Returns normalized list
  - If not fitted, print "Error: Not fitted!" and return None
- `is_ready()` - returns the __is_fitted status

**Formulas:**
- mean = sum(data) / len(data)
- std = sqrt(sum((x - mean)^2 for x in data) / len(data))

**Example:**
```python
preprocessor = DataPreprocessor()
preprocessor.fit([10, 20, 30, 40, 50])
result = preprocessor.transform([20, 40])
print(result)  # Normalized values
```
